```python
import sys
import os
sys.path.append(os.path.abspath('..'))

# Imports standards
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
import random
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Import de nos modules
from src.preprocessing import datagen   # maintenant ça fonctionne
from src.dataset_utils import create_split   # si besoin

%matplotlib inline
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 8)
```

```python
# Définir les chemins
train_dir = '../data/train'
val_dir = '../data/val'
test_dir = '../data/test'

# Vérifier si les dossiers existent, sinon les créer (optionnel)
if not os.path.exists(train_dir):
    print("Les dossiers de split n'existent pas. Création en cours...")
    create_split(
        src_dir='../data/raw',
        train_dir=train_dir,
        val_dir=val_dir,
        test_dir=test_dir,
        split_ratio=(0.7, 0.15, 0.15)
    )
else:
    print("Dossiers trouvés :", os.listdir('../data'))
```

```python
def count_images_per_class(base_dir):
    """
    Parcourt un dossier structuré en sous-dossiers (classes)
    et retourne un dictionnaire {classe: nombre_images}
    """
    class_counts = {}
    for cls in os.listdir(base_dir):
        cls_path = os.path.join(base_dir, cls)
        if os.path.isdir(cls_path):
            n_images = len([f for f in os.listdir(cls_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))])
            class_counts[cls] = n_images
    return class_counts

# Compter pour train, val, test
train_counts = count_images_per_class(train_dir)
val_counts = count_images_per_class(val_dir)
test_counts = count_images_per_class(test_dir)

print("Distribution train :", train_counts)
print("Distribution val :", val_counts)
print("Distribution test :", test_counts)
```

```python
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
plt.bar(train_counts.keys(), train_counts.values())
plt.xticks(rotation=90)
plt.title('Nombre d\'images par classe (train)')

plt.subplot(1,2,2)
plt.bar(val_counts.keys(), val_counts.values())
plt.xticks(rotation=90)
plt.title('Nombre d\'images par classe (validation)')
plt.tight_layout()
plt.show()
```

```python
# Choisir quelques classes au hasard (par exemple 5)
classes = list(train_counts.keys())
sample_classes = random.sample(classes, min(5, len(classes)))

fig, axes = plt.subplots(len(sample_classes), 4, figsize=(16, 4*len(sample_classes)))
if len(sample_classes) == 1:
    axes = [axes]

for i, cls in enumerate(sample_classes):
    cls_path = os.path.join(train_dir, cls)
    images = os.listdir(cls_path)
    # Prendre 4 images au hasard
    sample_imgs = random.sample(images, min(4, len(images)))
    for j, img_name in enumerate(sample_imgs):
        img_path = os.path.join(cls_path, img_name)
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        axes[i][j].imshow(img)
        axes[i][j].set_title(f'{cls}\n{img.shape}')
        axes[i][j].axis('off')
plt.tight_layout()
plt.show()
```

```python
# Fonction pour collecter les dimensions d'un échantillon d'images
def get_image_sizes(base_dir, sample_size=100):
    sizes = []
    classes = os.listdir(base_dir)
    for cls in classes:
        cls_path = os.path.join(base_dir, cls)
        if not os.path.isdir(cls_path):
            continue
        images = os.listdir(cls_path)
        # Prendre un échantillon par classe pour éviter de tout charger
        sample_imgs = random.sample(images, min(sample_size//len(classes), len(images)))
        for img_name in sample_imgs:
            img_path = os.path.join(cls_path, img_name)
            img = cv2.imread(img_path)
            if img is not None:
                sizes.append(img.shape[:2])  # (hauteur, largeur)
    return sizes

sizes = get_image_sizes(train_dir, sample_size=200)
heights, widths = zip(*sizes)

print(f"Hauteurs : min={min(heights)}, max={max(heights)}, moyenne={np.mean(heights):.1f}")
print(f"Largeurs : min={min(widths)}, max={max(widths)}, moyenne={np.mean(widths):.1f}")

# Visualisation
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.hist(heights, bins=20, alpha=0.7, label='Hauteurs')
plt.xlabel('Hauteur (pixels)')
plt.ylabel('Fréquence')
plt.legend()

plt.subplot(1,2,2)
plt.hist(widths, bins=20, alpha=0.7, color='orange', label='Largeurs')
plt.xlabel('Largeur (pixels)')
plt.ylabel('Fréquence')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
def check_channels(base_dir, sample_size=100):
    channels = []
    classes = os.listdir(base_dir)
    for cls in classes:
        cls_path = os.path.join(base_dir, cls)
        if not os.path.isdir(cls_path):
            continue
        images = os.listdir(cls_path)
        sample_imgs = random.sample(images, min(sample_size//len(classes), len(images)))
        for img_name in sample_imgs:
            img_path = os.path.join(cls_path, img_name)
            img = cv2.imread(img_path)
            if img is not None:
                channels.append(img.shape[2] if len(img.shape) == 3 else 1)
    unique, counts = np.unique(channels, return_counts=True)
    for u, c in zip(unique, counts):
        print(f"{c} images avec {u} canal/canaux")
    return channels

channels = check_channels(train_dir)
```

```python
# Charger une image exemple
ex_cls = random.choice(classes)
ex_img_name = random.choice(os.listdir(os.path.join(train_dir, ex_cls)))
ex_img_path = os.path.join(train_dir, ex_cls, ex_img_name)
img = cv2.imread(ex_img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = img.reshape((1,) + img.shape)  # ajouter dimension batch (1, H, W, C)

# Générer 8 images augmentées
plt.figure(figsize=(16,4))
for i in range(8):
    aug_img = next(datagen.flow(img, batch_size=1))[0].astype('uint8')
    plt.subplot(2,4,i+1)
    plt.imshow(aug_img)
    plt.axis('off')
plt.suptitle(f'Exemples d\'augmentation pour la classe {ex_cls}')
plt.show()
```

```python
def find_corrupted_images(base_dir):
    corrupted = []
    for cls in os.listdir(base_dir):
        cls_path = os.path.join(base_dir, cls)
        if not os.path.isdir(cls_path):
            continue
        for img_name in os.listdir(cls_path):
            img_path = os.path.join(cls_path, img_name)
            try:
                img = cv2.imread(img_path)
                if img is None:
                    corrupted.append(img_path)
            except:
                corrupted.append(img_path)
    return corrupted

corrupted_train = find_corrupted_images(train_dir)
corrupted_val = find_corrupted_images(val_dir)
corrupted_test = find_corrupted_images(test_dir)

print(f"Images corrompues dans train : {len(corrupted_train)}")
print(f"Images corrompues dans val : {len(corrupted_val)}")
print(f"Images corrompues dans test : {len(corrupted_test)}")

# Si il y en a, afficher les chemins
if corrupted_train:
    print("Exemples :", corrupted_train[:5])
```
